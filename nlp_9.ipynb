```python
from textblob import TextBlob


reviews = [
    "i lpve this product,it is amazing!",
    "this is tha worst purchase i ever made.",
    "the product is okay , not great."
]


for review in reviews:
  blob = TextBlob(review)
  polarity = blob.sentiment.polarity
  if polarity > 0:
    sentiment = "positive"
  elif polarity < 0:
    sentiment = "negative"
  else:
    sentiment = "neutral"

  print(f"Review: {review}\nSentiment: {sentiment}\nPolarity: {polarity}\n")
```
